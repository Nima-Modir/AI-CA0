```python
train['Age'] = train['Age'].fillna(int(train['Age'].mean()))
```

The Null datas of Age column are filled with the average of Non-Null datas of that column.
Disadvantage of filling with average : When there are lots of null datas in a cloumn, filling missing values with average may causes bias. However,the average value remains stable.
Advantages of filling with average : It makes the average value of that colmun remains stable and we can work with more datas when we fill the missing values.

```python
train['Fare']=(train['Fare']-train['Fare'].mean())/train['Fare'].std()
train['Parch']=(train['Parch']-train['Parch'].mean())/train['Parch'].std()
train['SibSp']=(train['SibSp']-train['SibSp'].mean())/train['SibSp'].std()
train['Age']=(train['Age']-train['Age'].mean())/train['Age'].std()
train['Sex']=(train['Sex']-train['Sex'].mean())/train['Sex'].std()
train['Pclass']=(train['Pclass']-train['Pclass'].mean())/train['Pclass'].std()
```

# 11

```python
train['Embarked']=train['Embarked'].astype('category').cat.codes
```

```python
train
```

```python
survived=train[train['Survived']==1]
unsurvived=train[train['Survived']==0]
```

```python
col = ['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']
for i in col :
    s1=survived[i].mean()
    s2=survived[i].std()
    x=np.linspace(-5,5,200)
    y=norm.pdf(x,s1,s2)
    plt.plot(x,y)
    s3=unsurvived[i].mean()
    s4=unsurvived[i].std()
    x=np.linspace(-5,5,200)
    y2=norm.pdf(x,s3,s4)
    plt.plot(x,y2)
    plt.title(i)
    plt.show()
    
```

I choose "PClass" as best variable for input of our model. In the PDF of "PClass" it can be seen that figures for survived and unsurvived conditions are seperatable and this variable is  more appropriate than others for our input.
Our predicting for surviving people is such that people (in test file) whose PClass is less than the intersection point in above figure(PClass) are survived and other are not.

```python
test=pd.read_csv('test.csv')
```

```python
test
```

```python
pclass_surv_mean=survived['Pclass'].mean()
pclass_surv_std=survived['Pclass'].std()
pclass_unsurv_mean=unsurvived['Pclass'].mean()
pclass_unsurv_std=unsurvived['Pclass'].std()
x=np.linspace(-5,5,200)
y1=norm.pdf(x,pclass_surv_mean,pclass_surv_std)
y2=norm.pdf(x,pclass_unsurv_mean,pclass_unsurv_std)
a=y2-y1
a[:50]=1
a[150:]=1
np.argmin(abs(a))
threshold=x[np.argmin(abs(a))]
```

```python
test['Pclass']=(test['Pclass']-test['Pclass'].mean())/test['Pclass'].std() #Normalization Pclass column
```

```python
test['Survived']=test['Pclass']<threshold
```

```python
test['Survived']=pd.get_dummies(test['Survived'],drop_first=True)
```

```python
test.to_csv('test_corrected.csv')
```
